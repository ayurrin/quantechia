```python
import pandas as pd
import numpy as np

def calculate_beta_from_df_fast(stock_returns: pd.DataFrame, benchmark_return: pd.DataFrame) -> pd.DataFrame:
    """
    Calculate the beta values of multiple stocks relative to a benchmark index.

    Parameters:
        stock_returns (pd.DataFrame): A DataFrame of stock closing returns. 
                                     Columns should be stock tickers, index should be datetime.
        benchmark_return (pd.DataFrame): A DataFrame containing the benchmark's closing return. 
                                        Must have a column named 'Close' and datetime index.

    Returns:
        pd.DataFrame: A DataFrame with columns ['ticker', 'beta'].
    """
    
    # Combine stock and benchmark returns into one DataFrame
    combined_returns = pd.concat([stock_returns, benchmark_return], axis=1).dropna(subset=benchmark_return.columns)

    # Compute covariance matrix (stocks + benchmark)
    cov_matrix = combined_returns.cov(ddof=0).to_numpy()

    # Calculate beta for each stock relative to the benchmark
    beta_values = cov_matrix[:-1, -1] / cov_matrix[-1, -1]

    # Create a DataFrame of betas
    beta_df = pd.DataFrame({
        'ticker': stock_returns.columns,
        'beta': beta_values
    })

    return beta_df



def calculate_beta_from_df(stock_returns: pd.DataFrame, benchmark_returns: pd.Series) -> pd.DataFrame:
    """
    Calculate beta values for multiple stocks relative to a benchmark using pairwise NaN removal.
    
    Parameters:
        stock_returns (pd.DataFrame): Daily returns of multiple stocks (tickers as columns).
        benchmark_returns (pd.Series): Daily returns of the benchmark index.
        
    Returns:
        pd.DataFrame: A dataframe with 'ticker' and 'beta' columns.
    """
    # Create an empty list to store betas
    beta_list = []

    # Loop through each stock (column)
    for ticker in stock_returns.columns:
        # Create a DataFrame for the stock and benchmark returns
        pair_df = pd.concat([stock_returns[ticker], benchmark_returns], axis=1).dropna()
        pair_df.columns = ['stock', 'benchmark']

        # Skip if not enough data (at least 2 data points required)
        if len(pair_df) < 2:
            continue

        # Calculate covariance matrix for this pair
        cov_matrix = pair_df.cov(ddof=0)

        # Calculate beta: Cov(stock, benchmark) / Var(benchmark)
        beta = cov_matrix.loc['stock', 'benchmark'] / cov_matrix.loc['benchmark', 'benchmark']

        # Append the result for this ticker
        beta_list.append({'ticker': ticker, 'beta': beta})

    # Return as a DataFrame
    return pd.DataFrame(beta_list)
```

```python

import pandas_datareader as web
def cal_price_data(date_s, date_e, ticker_list, bm_ticker='^TPX'):
    stock_price = web.DataReader(ticker_list, "stooq", start=date_s, end=date_e).sort_index()
    tp_price =  web.DataReader([bm_ticker], "stooq", start=date_s, end=date_e).sort_index()

    #beta
    # リターンを計算
    sp_returns = tp_price['Close'].pct_change().dropna()*100  # S&Pのリターン
    st_returns = stock_price['Close'].pct_change().dropna(how='all')*100  # 個別銘柄のリターン

    combined_returns = pd.concat([st_returns, sp_returns], axis=1).dropna(subset=[bm_ticker])
    cov_matrix = combined_returns.cov(ddof=0).to_numpy()  # 共分散行列を計算
    betas_matrix = cov_matrix[:-1, -1] / cov_matrix[-1, -1]  # ベータを計算
    beta = pd.Series(betas_matrix, index=st_returns.columns, name='beta')

    beta = beta.reset_index()
    beta.columns = ['ticker','Mkt']
    beta

    # 12か月前の価格を取得（252営業日 = 約12か月）
    price_12m_ago = stock_price.shift(252)

    # 1か月前の価格を取得（21営業日 = 約1か月）
    price_1m_ago = stock_price.shift(21)

    # 対数リターンを計算
    log_return = np.log(price_1m_ago / price_12m_ago)
    # 結果を表示

    log_rtn = log_return.dropna(how='all').resample('D').ffill()['Close'].stack().reset_index()
    log_rtn.columns = ['date', 'ticker', 'log_rtn']
    log_rtn['date'] = log_rtn['date'].astype(str)
    log_rtn.head()

    price_data = stock_price.resample('D').ffill()['Close'].stack().reset_index()
    price_data.columns = ['date', 'ticker', 'price']
    price_data['date'] = price_data['date'].astype(str)
    price_data.head()
    return beta, log_rtn, price_data

def make_std(data, v_weight, lower_tile=0.01, upper_tile=0.99, upper=3, lower=-3):
    data = pd.DataFrame(data)
    if len(data) > 100:
        lower_v = data.quantile(lower_tile).iloc[0]
        upper_v = data.quantile(upper_tile).iloc[0]
        data.clip(lower=lower_v, upper=upper_v, inplace=True)
    data.columns=v_weight.columns
    weighted_mean = (data * v_weight).sum() / v_weight.sum()
    weighted_std = (((data-weighted_mean) ** 2 * v_weight).sum() / v_weight.sum()) ** 0.5
    std_val = (data-weighted_mean)/weighted_std
    std_val.clip(lower, upper, inplace=True)
    return std_val

def cal_std(data, lower_tile=0.005, upper_tile=0.995, upper=3, lower=-3):
    data = pd.DataFrame(data)
    if len(data) > 100:
        # 上位1％と下位1％の値を各列ごとに計算
        lower_v = data.quantile(lower_tile)
        upper_v = data.quantile(upper_tile)
        data.clip(lower=lower_v, upper=upper_v, inplace=True, axis=1)

    std_val = (data-data.mean())/data.std()

    # クリップして上限下限を設定
    std_val.clip(lower=lower, upper=upper, inplace=True)
    std_val.clip(lower, upper, inplace=True)
    return std_val


def cal_fct_exp(data_raw):



    date_e = datetime.datetime.strptime(str(data_raw['date'].min()), '%Y-%m-%d')+datetime.timedelta(days=5)

    date_s = datetime.datetime.strptime(str(data_raw['date'].min()), '%Y-%m-%d')-datetime.timedelta(days=740)
    data_raw['ticker'] = data_raw['secCode'].astype(str).str[:4]+'.JP'

    ticker_list = list(data_raw['ticker'])
    beta, log_rtn, price_data =  cal_price_data(date_s, date_e, ticker_list)

    data = pd.merge(data_raw, beta, on=['ticker'],how='left')
    data = pd.merge(data, price_data, on=['date','ticker'],how='left')
    data = pd.merge(data, log_rtn, on=['date','ticker'],how='left')

    data['Size'] = np.log(data['shares_outstanding']*data['price'])
    data['1/PER'] = data['net_income_loss'] / (data['price'] * data['shares_outstanding'])
    data['ROA'] = data['net_income_loss'] / data['assets']
    data['ROE'] = data['net_income_loss'] / data['shareholders_equity']
    data['1/PBR'] = data['assets'] / (data['price'] * data['shares_outstanding'])
    data['asset_growth'] = (data['assets_b']-data['assets']) / data['assets']


    data['SMB'] = -cal_std(data['Size'])
    v_weight = data[['Size']]
    data['1/PBR_std'] = cal_std(data['1/PBR'])
    data['1/PER_std'] = cal_std(data['1/PER'])
    data['HML'] = make_std((data['1/PBR_std'] + data['1/PER_std'])/2,v_weight)

    data['ROA_std'] = cal_std(data['ROA'])
    data['ROE_std'] = cal_std(data['ROE'])
    data['RMW'] = make_std((data['ROA_std'] + data['ROE_std'])/2, v_weight)

    data['CMA'] = make_std(data['asset_growth'],v_weight)
    data['UMD'] = cal_std(data['log_rtn'])
    # Convert the DataFrame to JSON and return
    data = data.replace({np.nan: None, np.inf: None, -np.inf: None})
    # Reset the index and make the index a column
    
    df_reset = data[['secCode','date','filerName','EndDate','Mkt', 'HML', 'SMB', 'CMA', 'RMW', 'UMD']]
    # df_reset = data_raw[['secCode','date','filerName','EndDate']]

    # Convert to dictionary with index included
    result = df_reset.to_dict(orient='records')
    
    return result



def get_data(url):
    headers = {
        'User-Agent': 'email@gmail.com',
    }


    # リクエストを送信してJSONデータを取得
    response = requests.get(url, headers=headers)

    # レスポンスの内容を確認
    if response.status_code == 200:
        # JSONデータをDataFrameに読み込む
        data = response.json()
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
    return data

import requests

def cal_fct_exp_edgar(date, cik_list):

    
    # CIKコードのリストを取得（ここでは 'CIK' 列を持つと仮定）
    cik_list = df.iloc[:,0].tolist()
    print(cik_list)

    date_b = str(int(date[:4]) -1) + date[4:]

    url = "https://www.sec.gov/files/company_tickers.json"
    data = get_data(url)
    ticker_df = pd.DataFrame(data).T
    ticker_df.columns = ['cik','ticker','title']
    if cik_list==[]:

        # S&P 500構成銘柄がリストされているWikipediaページのURL
        url = 'http://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

        # Wikipediaページ中のすべての表を読み込む
        tables = pd.read_html(url)

        # S&P 500構成銘柄が含まれる最初の表を取得
        df_sp500 = tables[0]
        columns = [
            'Symbol', 'Security', 'GICS Sector', 'GICS Sub-Industry',
            'Headquarters Location', 'Date added', 'CIK', 'Founded'
        ]
        df_sp500.columns = columns

        sp_cik = df_sp500['CIK'].tolist()
    else:
        sp_cik = [int(i) for i in cik_list]
        ticker_df = ticker_df[ticker_df['cik'].isin(sp_cik)]
    concept_names = [
        'NetIncomeLoss',
        'Assets',
        # 'OperatingIncomeLoss',
        'StockholdersEquity',
        'CommonStockSharesOutstanding'
    ]
    concept_names_i = [
        'Assets',
        'StockholdersEquity',

    ]
    df = pd.DataFrame()

    for concept_name in concept_names:
        if concept_name == 'CommonStockSharesOutstanding':
            url = f'https://data.sec.gov/api/xbrl/frames/us-gaap/{concept_name}/shares/CY{date}I.json'
        elif concept_name in concept_names_i:
            url = f'https://data.sec.gov/api/xbrl/frames/us-gaap/{concept_name}/USD/CY{date}I.json'
        else:
            url = f'https://data.sec.gov/api/xbrl/frames/us-gaap/{concept_name}/USD/CY{date}.json'
        data = get_data(url)
        df_ = pd.DataFrame(data['data'])
        # df_ = df_[df_['cik'].isin(sp_cik)].loc[:,['cik','val']]
        df_ = df_.loc[:,['cik','end','val']]
        df_ = df_.rename(columns={'val':concept_name})
        if len(df) == 0:
            df = df_
        else:
            df = pd.merge(df, df_, on=['cik','end'])

    concept_name = 'Assets'
    url = f'https://data.sec.gov/api/xbrl/frames/us-gaap/{concept_name}/USD/CY{date_b}I.json'
    data = get_data(url)
    df_ = pd.DataFrame(data['data'])
    df_ = df_.loc[:,['cik','val']]
    df_ = df_.rename(columns={'val':concept_name})
    df = pd.merge(df, df_, on=['cik'], suffixes=['', '_b'])
    df_sp = df[df['cik'].isin(sp_cik)]
    data_raw = pd.merge(ticker_df,df_sp,  on=['cik'])
    data_raw = data_raw.rename(columns={'end':'date'})
    print(data_raw)

    date_e = datetime.datetime.strptime(str(df_sp['end'].max()), '%Y-%m-%d')+datetime.timedelta(days=5)

    date_s = datetime.datetime.strptime(str(df_sp['end'].min()), '%Y-%m-%d')-datetime.timedelta(days=740)


    ticker_list =  list(ticker_df[ticker_df['cik'].isin(df_sp['cik'].tolist())]['ticker'])
    beta, log_rtn, price_data =  cal_price_data(date_s, date_e, ticker_list, bm_ticker='^SPX')

    data = pd.merge(data_raw, beta, on=['ticker'],how='left')
    data = pd.merge(data, price_data, on=['date','ticker'],how='left')
    data = pd.merge(data, log_rtn, on=['date','ticker'],how='left')

    data['Size'] = np.log(data['CommonStockSharesOutstanding']*data['price'])
    data['1/PER'] = data['NetIncomeLoss'] / (data['price'] * data['CommonStockSharesOutstanding'])
    data['ROA'] = data['NetIncomeLoss'] / data['Assets']
    data['ROE'] = data['NetIncomeLoss'] / data['StockholdersEquity']
    data['1/PBR'] = data['Assets'] / (data['price'] * data['CommonStockSharesOutstanding'])
    data['asset_growth'] = (data['Assets_b']-data['Assets']) / data['Assets']


    data['SMB'] = -cal_std(data['Size'])
    v_weight = data[['Size']]
    data['1/PBR_std'] = cal_std(data['1/PBR'])
    data['1/PER_std'] = cal_std(data['1/PER'])
    data['HML'] = make_std((data['1/PBR_std'] + data['1/PER_std'])/2,v_weight)

    data['ROA_std'] = cal_std(data['ROA'])
    data['ROE_std'] = cal_std(data['ROE'])
    data['RMW'] = make_std((data['ROA_std'] + data['ROE_std'])/2, v_weight)

    data['CMA'] = make_std(data['asset_growth'],v_weight)
    data['UMD'] = cal_std(data['log_rtn'])
    data = data.replace({np.nan: None, np.inf: None, -np.inf: None})

    # df_reset = data_raw[['secCode','date','filerName','EndDate']]

    # Convert to dictionary with index included
    result = data[['cik','date','ticker','title','Mkt', 'HML', 'SMB', 'CMA', 'RMW', 'UMD']].to_dict(orient='records')
    
    return result
```
